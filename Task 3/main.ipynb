```python
from sklearn.preprocessing import LabelEncoder
label_encoder = LabelEncoder()
iris_data['species'] = label_encoder.fit_transform(iris_data['species'])
```

**Step 6: Creating prediction Model**

1. Split the data in terms of independent variable and dependent variable.  
2. Split the data into training and testing data
3. Train the model using Linear regression
4. Check accuracy

```python
#Step 1: Splitting data into independent and dependent variable
#As we see in correlation map, price only depends upon sepal_length and petal_length so
#the code is as follows:
# Selecting columns for prediction
selected_columns = ['sepal_length', 'petal_length', 'species']
selected_data = iris_data[selected_columns]
# Split the data into features (X) and target variable (y)
X = selected_data[['sepal_length', 'petal_length']]
Y = selected_data['species']
```

```python
X.head()
```

```python
Y.head()
```

```python
#Step 2: Splitting the data into training and testing data set.
#We know best split method is 80 - 20 rule.
X_train, X_test, Y_train, Y_test = train_test_split(X,Y, test_size=0.3, random_state=42)
```

```python
X_train.head()
```

```python
X_test.head()
```

```python
Y_train.head()
```

```python
Y_test.head()
```

```python
#Step 3: Linear Regression Model

#Import libraries
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score

#Create a model
model = LinearRegression()
model.fit(X_train, Y_train)
```

```python
#Step 4: Make predictions on the test set
y_pred = model.predict(X_test)
```

```python
#Step 5: Evaluate the model
from sklearn.metrics import mean_squared_error, r2_score
mse = mean_squared_error(Y_test, y_pred)
r2 = r2_score(Y_test, y_pred)
print(f'Mean Squared Error: {mse:.2f}')
print(f'R-squared Value: {r2:.2f}')
```

```python
#Step 6 Plot the predicted vs. actual values
plt.figure(figsize=(10, 6))
plt.scatter(Y_test, y_pred)
plt.plot([min(Y_test), max(Y_test)], [min(Y_test), max(Y_test)], linestyle='--', color='red', linewidth=2)
plt.title('Actual vs. Predicted Iris prediction')
plt.xlabel('Actual Car Iris prediction')
plt.ylabel('Predicted Iris prediction')
plt.show()
```

```python
# As this model is not suitable so we will go with logistic regression now.
# Create a logistic regression model and train it
model1 = LogisticRegression()
model1.fit(X_train, Y_train)
```

```python
# Make predictions on the test set
y_pred = model1.predict(X_test)
```

```python
# Evaluate the model
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
accuracy = accuracy_score(Y_test, y_pred)
conf_matrix = confusion_matrix(Y_test, y_pred)
classification_rep = classification_report(Y_test, y_pred)
```

```python
print(f'Accuracy: {accuracy:.2f}')
print('Confusion Matrix:')
print(conf_matrix)
print('Classification Report:')
print(classification_rep)
```

```python
#Again Plot the predicted vs. actual values
plt.figure(figsize=(10, 6))
plt.scatter(Y_test, y_pred)
plt.plot([min(Y_test), max(Y_test)], [min(Y_test), max(Y_test)], linestyle='--', color='red', linewidth=2)
plt.title('Actual vs. Predicted Iris prediction')
plt.xlabel('Actual Iris prediction')
plt.ylabel('Predicted Iris prediction')
plt.show()
```
